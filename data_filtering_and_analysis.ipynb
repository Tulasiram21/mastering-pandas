```python
import pandas as pd
```

```python
df=pd.read_csv('vgsales.csv')
df.head()
```

```python
df.info()
```

```python
df_cleaned=df.dropna(subset=['Year'])
```

```python
df_cleaned.info()
```

## Filtering

```python
nintendo_games=df_cleaned[df_cleaned['Publisher']=='Nintendo']
nintendo_games
```

```python
recent_games=df_cleaned[df_cleaned['Year']>=2010]
recent_games
```

```python
nintendo_recent_games=df_cleaned[(df_cleaned['Publisher'] == 'Nintendo') & (df_cleaned['Year'] >= 2010) ]

nintendo_recent_games
```

## sorting the data

```python
df_cleaned.sort_values(by='Year')
```

```python
df_cleaned.sort_values(by='Year',ascending=False)
```

## Grouping the Data

```python
# avg sales for each genre

df_cleaned.groupby('Genre')['Global_Sales'].mean()
```

```python
# top 10 best-selling games by Global_Sales

best_selling_games=df_cleaned.sort_values(by='Global_Sales',ascending=False)
best_selling_games.head(10)
```

```python
# total Global_Sales for each Genre

df_cleaned.groupby('Genre')['Global_Sales'].sum()
```

```python
# average Global_Sales for each Platform

df_cleaned.groupby('Platform')['Global_Sales'].mean()
```
