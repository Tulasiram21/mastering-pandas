```python
import pandas as pd
```

```python
df=pd.read_csv('ADANIPORTS.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Decade']=(df['Date'].dt.year//10)*10
```

```python
df.info()
```

# Extracting Day from Date column

```python
df['Day']=df['Date'].dt.day
```

# Extracting Month from Date Column

```python
df['Month']=df['Date'].dt.month
```

# Extracting Year from Date Column

```python
df['Year']=df['Date'].dt.year
```

# Extracting Day name from Date Column

```python
df['Day Name']=df['Date'].dt.day_name()
```

```python
df.head()
```

# The average Open price for each day of the week

```python
df.groupby(df['Day Name'])['Open'].mean()
```

# Extracting the Day of week as a number(eg., 0 for Monday, 6 for Sunday)

```python
df['Day of Week']=df['Date'].dt.dayofweek
```

# EXtracting Quarter of the year(e.g., 1 for Jan-Mar)

```python
df['Quarter']=df['Date'].dt.quarter
```

# .dt.is_leap_year: 

Returns True if the year is a leap year, False otherwise.

```python
df['Leap Year']=df['Date'].dt.is_leap_year
```
