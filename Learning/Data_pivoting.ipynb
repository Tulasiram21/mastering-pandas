```python
import pandas as pd
```

```python
df=pd.read_csv('vgsales.csv')
```

```python
df.head()
```

```python
df.info()
```

## Creating Pivot Table

```python
pivot_table=df.pivot_table(index='Platform',columns='Genre',values='Global_Sales')
pivot_table
```

## agggfunc : BY default pivot_table uses mean()

```python
pivot_table=df.pivot_table(index='Platform',columns='Genre',values='Global_Sales',aggfunc=sum)
```

```python
pivot_table
```

```python
pivot_table=df.pivot_table(index='Platform',columns='Genre',values='Global_Sales',aggfunc='count')
```

```python
pivot_table
```
